```python
import pandas as pd
import numpy as np
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset["sex_male"] = dataset["sex_male"].astype(int)

dataset["smoker_yes"] = dataset["smoker_yes"].astype(int)
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.3,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svr', SVR())
])

param_grid={'svr__kernel':['rbf','poly','sigmoid','linear'],
            'svr__C':[10,100,1000,2000,3000],'svr__gamma':['auto','scale']}

grid=GridSearchCV(pipeline,param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(X_train,Y_train)
```

```python
re=grid.cv_results_
print("The R_Score value for best parameter is {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```
